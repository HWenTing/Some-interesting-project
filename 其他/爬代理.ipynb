```python
import urllib
import re
import requests
import time
from threading import Thread
from threading import Lock
```

#### 获取代理

```python

def get_all(n):
    headers = {'User-Agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/52.0.2743.116 Safari/537.36'}
    li = []
    for i in range(n):
        #从xici网站的高匿页面获取ip
        url='http://www.xicidaili.com/nn/%d'%(i+1)
        r = requests.get(url,headers=headers)
        data=r.text
        #抓取所需数据的正则表达式
        p=r'<td>(.*?)</td>\s+<td>(.*?)</td>\s+<td>\s+(.*?)\s+</td>\s+<td class="country">(.*?)</td>'
        find_list=re.findall(p,data)
    #将ip地址与端口组成规定格式
        for row in find_list:
            ip=row[0]+':'+row[1]
            li.append(ip)
    return li
```

```python
list_proxies = get_all(1)
```

#### 根据得到的代理ip，设置proxy的格式

```python
def get_proxy(checking_ip):
    #根据得到的代理ip，设置proxy的格式
    proxy_ip = 'http://' + checking_ip
    proxy_ips = 'https://' + checking_ip
    proxy = {'https': proxy_ips, 'http': proxy_ip}
    return proxy
```

#### 验证代理,返回可用代理

```python
def check_ip(url_list,list_proxies): 
    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/61.0.3163.100 Safari/537.36'}
    success_list = []
    for i in list_proxies:
        proxy=get_proxy(i) 
        try:
            print(i)
            for j in url_list:
                page = requests.get(j, headers=headers, proxies=proxy,timeout=1)   
        except:
            print(i,'失败')
        else:
            print(i,'成功')
            success_list.append(i)
    return success_list
```
