```python
import pandas as pd
import numpy as np
import collections #统计
from wordcloud import WordCloud,ImageColorGenerator #分词
import matplotlib.pyplot as plt #作图
```

```python
with open(r'C:/Users/HP/新建文件夹/pa_chong/poem/朝代/南北朝.txt',encoding='utf-8') as f:
    line = f.readline()
words=list(line)
```

```python
# 词频统计
mycount = collections.Counter(words)
for key, val in mycount.most_common(10):  # 有序（返回前10个）
    print(key, val)
```

```python
# 生成词云
font_path = "D:\Fonts\simkai.ttf"#字体路径
wc = WordCloud(
    background_color="white", #背景颜色
    max_words=1000, #显示最大词数
    font_path=font_path,
    min_font_size=5,
    max_font_size=100, 
    mask=plt.imread("poem2.jpg"),
)
wc.generate_from_frequencies(mycount)
# image_colors = ImageColorGenerator(back_color)
plt.figure()
# plt.imshow(wc.recolor(color_func=image_colors))
plt.imshow(wc)
plt.axis('off')
plt.savefig('heart2.jpg', dpi=1600) 
print('生成词云成功!')
```
