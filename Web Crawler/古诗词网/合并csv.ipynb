```python
# coding: utf-8
import pandas as pd
import glob
#获取文件夹下csv文件
csv_files = glob.glob(r'C:/Users/HP/新建文件夹/pa_chong/poem/*/*.csv')
path = 'total.csv'
content = pd.DataFrame(columns=['author','dynasty','title','content'])
content.to_csv(path,index=False,encoding="utf_8_sig")
count=0
for inputfile in csv_files:
    print(str(count)+'    '+inputfile)
    f=open(inputfile,'rb')
    content = pd.read_csv(f, header=None)[1:]
    content.to_csv(path,index=False,mode='a', header=False,encoding="utf_8_sig")
    f.close()
    count= count+1
```
